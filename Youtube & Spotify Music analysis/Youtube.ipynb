#### **Data Preprocessing**

**Handling missing values**
##### Impute missing values in Spotify features with the median

```python
spotify_features = ['Danceability', 'Energy', 'Key', 'Loudness', 'Speechiness', 
                    'Acousticness', 'Instrumentalness', 'Liveness', 'Valence', 'Tempo', 'Duration_ms']
for feature in spotify_features:
    df[feature].fillna(df[feature].median(), inplace=True)
```

**Dropping rows where 'Stream' is missing**

```python
df = df.dropna(subset=['Stream'])
```

**Convert 'Duration_ms' from milliseconds to minutes**

```python
df.loc[:, 'Duration_min'] = df['Duration_ms'] / 60000
```

**Drop 'Unnamed: 0'**

```python
if 'Unnamed: 0' in df.columns:
    df.drop('Unnamed: 0', axis=1, inplace=True)
```

**1. Which artists have the most viewed music videos on YouTube?**

```python
# Summing up views by artist and sorting
artist_views = df.groupby('Artist')['Views'].sum().sort_values(ascending=False).head(10)
# Bar chart to display top artists by views
artist_views.plot(kind='bar')
plt.title('Top 10 Artists with Most Viewed Music Videos on YouTube')
plt.xlabel('Artist')
plt.ylabel('Total Views')
plt.show()
```

The bar chart provides a clear ranking of the top 10 artists based on 
the total views of their music videos on YouTube. The highest bars 
represent artists with the greatest viewership, indicating a strong fan 
base or wide appeal on the platform. This visualization helps in 
identifying the most popular artists in terms of YouTube views, 
highlighting those who have successfully captured the audience's 
attention.

```python
# Boxplot for views based on duration category
sns.boxplot(x='Duration_Category', y='Views', data=df)
plt.title('Song Popularity by Duration - Views')
plt.show()
```
